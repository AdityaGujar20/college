```python
import numpy as np
import matplotlib.pyplot as plt

# Function to compute DTFT (approximation using dense DFT points)
def compute_dtft(x, N=2048):
    X = np.fft.fftshift(np.fft.fft(x, N))
    w = np.linspace(-np.pi, np.pi, N)
    return w, np.abs(X)

# Function to compute DFT
def compute_dft(x, N):
    X = np.fft.fft(x, N)
    f = np.arange(N)
    return f, np.abs(X)

# Sampling parameters
fs = 1000
t = np.arange(0, 1, 1/fs)  # 1 second duration
```

```python
# Parameters
f = 5  # 5 Hz
x = np.sin(2 * np.pi * f * t)

# Time-domain plot
plt.figure()
plt.plot(t[:200], x[:200])  # show first 200 samples
plt.title("Time Domain - Sinusoidal (5 Hz)")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()

# DTFT
w, X_dtft = compute_dtft(x)
plt.figure()
plt.plot(w, X_dtft)
plt.title("DTFT - Sinusoidal")
plt.xlabel("Frequency (rad/sample)")
plt.ylabel("|X(w)|")
plt.grid(True)
plt.show()

# DFT
f_dft, X_dft = compute_dft(x, 1024)
plt.figure()
plt.stem(f_dft[:512], X_dft[:512])
plt.title("DFT - Sinusoidal")
plt.xlabel("Frequency bin")
plt.ylabel("|X[k]|")
plt.grid(True)
plt.show()
```

```python
# Composite signal: 5 Hz + 50 Hz
x1 = np.sin(2 * np.pi * 5 * t)
x2 = 0.5 * np.sin(2 * np.pi * 50 * t)
x_comp = x1 + x2

# Time-domain
plt.figure()
plt.plot(t[:200], x_comp[:200])
plt.title("Time Domain - Composite Signal (5 Hz + 50 Hz)")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()

# DTFT
w, X_dtft = compute_dtft(x_comp)
plt.figure()
plt.plot(w, X_dtft)
plt.title("DTFT - Composite Signal")
plt.xlabel("Frequency (rad/sample)")
plt.ylabel("|X(w)|")
plt.grid(True)
plt.show()

# DFT
f_dft, X_dft = compute_dft(x_comp, 1024)
plt.figure()
plt.stem(f_dft[:512], X_dft[:512])
plt.title("DFT - Composite Signal")
plt.xlabel("Frequency bin")
plt.ylabel("|X[k]|")
plt.grid(True)
plt.show()
```

```python
# Exponentially decaying signal
a = 0.05
x_exp = np.exp(-a * t)

# Time-domain
plt.figure()
plt.plot(t[:200], x_exp[:200])
plt.title("Time Domain - Exponentially Decaying Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()

# DTFT
w, X_dtft = compute_dtft(x_exp)
plt.figure()
plt.plot(w, X_dtft)
plt.title("DTFT - Exponential Signal")
plt.xlabel("Frequency (rad/sample)")
plt.ylabel("|X(w)|")
plt.grid(True)
plt.show()

# DFT
f_dft, X_dft = compute_dft(x_exp, 1024)
plt.figure()
plt.stem(f_dft[:512], X_dft[:512])
plt.title("DFT - Exponential Signal")
plt.xlabel("Frequency bin")
plt.ylabel("|X[k]|")
plt.grid(True)
plt.show()
```

```python
# Rectangular pulse (width 100 samples)
x_rect = np.zeros(len(t))
x_rect[200:300] = 1

# Time-domain
plt.figure()
plt.plot(t[:400], x_rect[:400])
plt.title("Time Domain - Rectangular Pulse")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()

# DTFT
w, X_dtft = compute_dtft(x_rect)
plt.figure()
plt.plot(w, X_dtft)
plt.title("DTFT - Rectangular Pulse")
plt.xlabel("Frequency (rad/sample)")
plt.ylabel("|X(w)|")
plt.grid(True)
plt.show()

# DFT
f_dft, X_dft = compute_dft(x_rect, 1024)
plt.figure()
plt.stem(f_dft[:512], X_dft[:512])
plt.title("DFT - Rectangular Pulse")
plt.xlabel("Frequency bin")
plt.ylabel("|X[k]|")
plt.grid(True)
plt.show()
```
